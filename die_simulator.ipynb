Simulate a fair die and a biased 6-sided die. 
The biased die has probabilities {0.15, 0.15, 0.15, 0.15, 0.15, 0.25}. 
Create a visualization that compares outcomes of multiple rolls of

The user of your notebook should be able to alter the number of simulations as an argument to a function.â€‹

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Fair die: each side has the same probability of rolling up, each side has probability of 1/6th 
#Biased 6-sided die: Probability of landing each face are not the same. The probability for each side are as follows
#{0.15, 0.15, 0.15, 0.15, 0.15, 0.25}

#function to simulate fair and biased die rolls 
#fair_die_num = number of times fair die is rolled
#biased_die_num = number of times biased die is rolled 
def dieSimulation(fair_die_num, biased_die_num):
    #creating array of probabilities for fair and biased die
    fair_prob = np.ones(6, dtype = int) / 6 #Equal probability for each side 
    biased_prob = np.array([0.15, 0.15, 0.15, 0.15, 0.15, 0.25]) #given probabilities for each side 

    #using Numpy random choice function to return random array based on each die's probability. returns an array of random samples  
    #reference - https://sparkbyexamples.com/python/how-to-use-numpy-random-choice-in-python/ 
    fair_rolls = np.random.choice(6, size = fair_die_num, p = fair_prob) + 1 #adding one to allow numbers from 0 to 6, rather than 5 
    biased_rolls = np.random.choice(6, size = biased_die_num, p = biased_prob) + 1 #adding one to allow numbers from 0 to 6, rather than 5 

    #returning two numpy arrays 
    return fair_rolls, biased_rolls
    
#user entering the number of simulations
fair_value = input("Enter the amount of time you would like to roll the fair die: ")
biased_value = input("Enter the amount of time you would like to roll the biased die: ")

#calling siumlation function based on user specified values 
fair_results, biased_results = dieSimulation(int(fair_value), int(biased_value))

print(fair_results)
print(biased_results)
```

```python
# Histogram for fair die
plt.hist(fair_results, bins=np.arange(0.5, 7.5,1))
plt.title("Fair Die Rolls")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
```

```python
# Histogram for fair die
plt.hist(biased_results, bins=np.arange(0.5, 7.5, 1))
plt.title("Biased Die Rolls")
plt.xlabel("Outcome")
plt.ylabel("Frequency")
```
